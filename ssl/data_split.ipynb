```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
BASE_PATH = Path("../data/")

TEST_DATA_PATH = BASE_PATH / "test_data.csv"
TRAIN_DATA_PATH = BASE_PATH / "train_data.csv"
UNLABELLED_DATA_PATH = BASE_PATH / "unlabelled_data.csv"

NEW_PATH = BASE_PATH / 'sequential'
NEW_TEST_DATA_PATH = NEW_PATH / "test_data.csv"
NEW_VAL_DATA_PATH = NEW_PATH / "val_data.csv"
NEW_TRAIN_DATA_PATH = NEW_PATH / "train_data.csv"
NEW_UNLABELLED_DATA_PATH = NEW_PATH / "unlabelled_data.csv"
```

```python
test_df = pd.read_csv(TEST_DATA_PATH)
train_df = pd.read_csv(TRAIN_DATA_PATH)
unlabelled_df = pd.read_csv(UNLABELLED_DATA_PATH)
```

```python
test_df['auto_id'] = range(1, len(test_df) + 1)
print(len(test_df))
test_df.head()
```

```python
train_df['auto_id'] = range(1000, 1000 + len(train_df))
print(len(train_df))
train_df.head()
```

```python
unlabelled_df['auto_id'] = range(2000, 2000 + len(unlabelled_df))
print(len(unlabelled_df))
unlabelled_df.head()
```

```python
#full_X_train = train_df.drop(columns=['Label'])
#full_y_train = train_df['Label']
```

Proportion: 80% / 20%

```python
# Split the train_data into training and validation sets
new_train_df, val_df = train_test_split(
    train_df,
    test_size=0.2,
    random_state=42,
    stratify=train_df['Label'])

print(f'Training data shape: {new_train_df.shape}')
print(f'Validation data shape: {val_df.shape}')
```

```python
test_df.to_csv(NEW_TEST_DATA_PATH, index=False)
new_train_df.to_csv(NEW_TRAIN_DATA_PATH, index=False)
val_df.to_csv(NEW_VAL_DATA_PATH, index=False)
unlabelled_df.to_csv(NEW_UNLABELLED_DATA_PATH, index=False)
```
